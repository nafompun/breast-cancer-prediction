```python
# Importing the necessary packages for plotting the violin plot. 
import numpy as np 
import pandas as pd 
import seaborn as sns   
import matplotlib.pyplot as plt
import time
from subprocess import check_output
```

```python
col = bc.columns       
print(col)
```

```python
# y includes our labels and x includes our features
y = bc.diagnosis                         
list = ['id','diagnosis']
x = bc.drop(list,axis = 1 )
x.head()
```

```python
# first ten features distribution using violin plot
bc_dia = y
bc = x
bc_n_2 = (bc - bc.mean()) / (bc.std())              
bc = pd.concat([y,bc_n_2.iloc[:,0:10]],axis=1)
bc = pd.melt(bc,id_vars="diagnosis",var_name="features",value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value",hue = 'diagnosis', data=bc,split=True, inner="quart")
plt.xticks(rotation=90)
plt.title('Representation of first 10 features')
```

```python
# second ten features distribution using violin plot
bc = pd.concat([y,bc_n_2.iloc[:,10:20]],axis=1)
bc = pd.melt(bc,id_vars="diagnosis",var_name="features",value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value",hue = 'diagnosis', data=bc,split=True, inner="quart")
plt.xticks(rotation=90)
plt.title('Representation of second 10 features')
```

```python
# Third ten features distribution using violin plot

bc = pd.concat([y,bc_n_2.iloc[:,20:31]],axis=1)
bc = pd.melt(bc,id_vars="diagnosis",var_name="features",value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value",hue = 'diagnosis', data=bc,split=True, inner="quart")
plt.xticks(rotation=90)
plt.title('Representation of features using boxplot')
```

```python
#Boxplot representation of features. 
plt.figure(figsize=(10,10))
sns.boxplot(x="features", y="value", hue="diagnosis", data=bc)
plt.xticks(rotation=90)
plt.title('Represntation of features using boxplot')
```
